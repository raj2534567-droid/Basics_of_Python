# Types of operators

```python
# Types of operators
# Arithmatic Operator
# Relational/ Comparisonal Operator
# Logical Operator
# Membership Operator



# Arithmatic Operator (+ , - , *, / , ** , % , //)

a = 4
b = 2

print(a+b)
print(a-b)
print(a/b)   # it always gives in float
print(a%b)    # it gives reminder
print(a*b)
print(a**b)
print(a//b)   # complete divisiom or always gives integer value
```

```python
# Relational Operator / Comparisonal ( == , != , > , < , >= , <=)
# it return values in boolean data type
a = 50
b = 20

print(a == b)  # isequal sign
print(a != b)   #  notequal sign
print(a>b)       # greater to sign
print(a<b)       # smaller to sign 
print(a>=b)       # greater is equal to sign
print(a<=b)       # smaller is to equal sign
```

```python
# Assigment operator (= , += , -= , *= , /= , %= , **=)
# it works assigning the value to variable

a = 10
a += 10  #  a = a + 10
a -= 10 # a = a - 10
a *= 2 
a /= 2
a **= 2
a //= 2
print(a)
```

```python
# Logical operators (not , or , and)
# not it  always gives us the opposite values 

a = 3
b = 1

print(not(a>b))

# and it gives true values if both condition is true

print((a>b) and (b<a)) # both condition is getting true here
print((a>b) and (a<b)) # only condition is ture here and one is getting false

# or it gives true value if one condition will be true

print((a>b) or (a < b)) # so it will gives the true value
```

```python
# Membership operators
# The membership operators in python checks if an elements is present in a collection.


a = [1,2,3,4,5,6]

print(2 in a)  # this is printing true 

print(7 in a)  # this is printing false because 7 is not in the list
```
