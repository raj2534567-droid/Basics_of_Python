# Types of data (Primary data types, Built-in data types)

```python
# there are 6 types of built in data type:
# Implicit data type ( primary data type):

# integer = int
# float = float
# string = str
# boolean = bool
# None = None
# complex = complex



Age = 34             # int
Marks = 98.7         # float
Name = 'Rajeshwari'  # string
World = True         # Boolean
Time = None          # None
B = 3+5j             # Complex



# We can also check the data type of these variables:


print(type(Age))
print(type(Marks))
print(type(Name))
print(type(World))
print(type(Time))
print(type(B))
```
