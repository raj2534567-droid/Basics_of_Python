# Type conversion

```python
# there are two type of type conversion
# implicit and explicit(type casting)
# implicit : python automatically convert data type

# Example:
a = 3
b = 2.5

print(a+b)  # it automatically convereted a in float

```

```python
# explicit : manually convert data type

a = int("2")
b = 4.25

print(a+b)

```

```python
a = 1

a = bool(a)  # converting in bool

print(a)
```

```python
a = str(6)  # converting integer to string
b = "abcd"

print(a+b) 
```
