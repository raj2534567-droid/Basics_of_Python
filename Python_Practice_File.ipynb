```python
a=int(input("enter 1st number"))
b=int(input("enter 2nd number"))
c=int(input("enter 3rd number"))
if a>b :
    print("a is greatest")
elif b>c and b>a : 
    print("b is greatest")
else :
    print("c is greatest")
```

```python
# Nested if
a=int(input("enter 1st number"))
b=int(input("enter 2nd number"))
c=int(input("enter 3rd number"))
if a>b :
    if a>c :
        print("a is greatest")
elif b>c :
    print("b is greatest")
else :
    print("c is greatest")   
```

```python
# Nested if , if with 4 conditions and/or
a=int(input())
b=int(input())
c=int(input())
d=int(input())
if a>b and a>c and a>d : #and : all condition should br ture
    print("A is greatest")
elif b>a and b>c and b>d :
    print("B is greatest")
elif c>a and c>b and c>d :
    print("C is greatest")
else :
    print("D is greatest")
```

```python
# Nested if :
a=int(input())
b=int(input())
c=int(input())
d=int(input())
if a>b :
    if a>c :
        if a>d :
            print("a is greatest")
elif b>c:
    if b>d :
        print("b is greatest")
elif c>d :
    if c>a :
        if c>b :
            print("c is greatest")
else :
    print("d is greatest")
```
