# string and conditional formating

```python
# string string is a data type that store sequence of character
# whatever we write inside "" these inverted comma called string
# we can write string in three ways ('' , "" , """ """)
# string are immutable we cannot add or remove element in original string
# when we use string methods the result shows us in new string
# change has not made in original string 
```

```python
# basic operation in string

# concatenation : it add the two string & gives single result


a = "Patel"
b = "Rajeshwari"
c = a+b

print(c)

# len : it count the lenghts

print(len(c))

```

```python
# indexing : indexing is like a position of numbers
# it always starts with 0

A = 'Rajeshwari'

print(A[0])
print(A[3])
print(A[9])

# WE can only access the string with the help of indexing
# We cannot replace or assing the new values

# slicing : Accesing part of astring to get the substring 
# strarting index included, ending index is not included

A = 'Rajeshwari'

print(A[0:6])
print(A[6:])
print(A[:4])
print(A[::2])  # starting, ending, steps
print(A[::-1])  # neagative slicing


# Neagativ indexing

A = 'Rajeshwari'

print(A[-5])
print(A[-9])

# Neagative slicing

print(A[-4:])
      
```

```python
# Methods of string

A = 'rajeshwari'

print(A.endswith('i'))

print(A.capitalize())

print(A.replace('a','o'))
print(A.replace('rajeshwai','shwari'))

print(A.find("s")) # it tells us index number of first occurance
print(A.find("t")) # it will return -1 not error

print(A.count('a')) # it tells us the count of word
```

# Conditional formating

```python
# conditional formating : if , elif , else
# conditional formating are use in programing to make precise desicion

# Example:

Age = 16

if Age >= 18:
    print("can vote")
else:
    print("cannot vote")
    
```

```python
# if else elif
# befor using elif there must to give a if
# we can use if and multiple elif together 

light = "red"

if(light == "red"):
    print("stop")
elif(light == "green"):
    print("go")
elif(light == "yellow"):
    print("wait")
else:
    print("light is broken")

# in the condition chechk  the condition and print firt true values
```

```python
# nested if

age = 62

if age >= 18:
    if age >= 60:
        print('cannot drive')
    else:
        print('can drive')
    
```
